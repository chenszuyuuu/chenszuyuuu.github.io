```python
import pandas as pd
```

```python
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
us = df[df["country"]==217]
us.head()
```

```python
us = us[["eventid", "iyear", "imonth", "iday", "nkill","country", "country_txt", "provstate", "city", "success", "gname", "weaptype1", "weaptype1_txt", "weapdetail"]]
```

```python
us = us.reset_index(drop=True)
```

```python
us.isnull().sum()
```

```python
us.info()
```

```python
us.nkill.isnull().sum()
```

```python
groupby = us[["eventid", "iyear"]].groupby(["iyear"]).count()
groupby.to_csv("terr_year.csv")
```

```python
groupby[groupby["iyear"]==1993]
```

```python
index_l = []
year_l = []
month_l = []
day_l = []
kill_l = []
state_l = []
city_l = []
success_l = []
terr_l = []
weap_l = []
weapdetail_l = []
for i in range(len(us)):
    index = int(i)
    year = int(us["iyear"][i])
    month = int(us["imonth"][i])
    day = int(us["iday"][i])
    if pd.isna(us["nkill"][i])==True:
        kill = "null"
    else:
        kill = int(us["nkill"][i])
    if pd.isna(us["weapdetail"][i])==True:
        weapdetail = "null"
    else:
        weapdetail = us["weapdetail"][i]
    if pd.isna(us["provstate"][i]) == True:
        state = "null"
    else: 
        state = us["provstate"][i]
    city = us["city"][i]
    success = int(us["success"][i])
    terr_group = us["gname"][i]
    weaptype = us["weaptype1_txt"][i]
    
    index_l.append(index)
    year_l.append(year)
    month_l.append(month)
    day_l.append(day)
    kill_l.append(kill)
    success_l.append(success)
    state_l.append(state)
    city_l.append(city)
    terr_l.append(terr_group)
    weap_l.append(weaptype)
    weapdetail_l.append(weapdetail)
```

```python
index_l.append(-1)
year_l.append(1993)
month_l.append("null")
day_l.append("null")
kill_l.append("null")
success_l.append("null")
state_l.append("null")
city_l.append("null")
terr_l.append("null")
weap_l.append("null")
weapdetail_l.append("null")
```

```python
dic = {
    "index": index_l,
    "year": year_l,
    "month": month_l,
    "day": day_l,
    "success_or_not": success_l,
    "state": state_l,
    "city": city_l,
    "kill": kill_l,
    "terr_group": terr_l,
    "weapon": weap_l,
    "weapdetail": weapdetail_l
}
dic = pd.DataFrame.from_dict(dic)
```

```python
import json

json = dic.to_json(orient='records')
with open('terr.json', 'w') as f:
    f.write(json)
```
